```python
props = catalog.load("properties_project_CHIP1A")
data = catalog.load(
    "data_project_CHIP1A",
)
```

```python
df = data[props.loc[0, "data_key"]]()
```

```python
def get_forward(df):
    forward = df["Vg (V)"].diff() > 0
    return df[forward | forward.shift(-1, fill_value=False)]


def get_backward(df):
    dv = df["Vg (V)"].diff()
    mask = (dv < 0) | (dv.shift(-1) < 0)
    out = df[mask]
    return out.sort_values(by="Vg (V)")
```
